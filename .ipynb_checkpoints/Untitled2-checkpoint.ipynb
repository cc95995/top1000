```python
enterprise = '中国联合网络通信集团有限公司'
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
data1 = pd.read_csv("D:/pytest/liushui/1liushui.csv", encoding="gbk")
data1_lt = data1[data1['企业名称系统存储'] == enterprise]
for i in range(2,12):
    data2 = pd.read_csv("D:/pytest/liushui/%dliushui.csv"%i, encoding="gbk")
    data2_lt = data2[data2['企业名称系统存储'] == enterprise]
    data = pd.concat([data1_lt, data2_lt])
    data1_lt = data 
```

```python
data1_lt_copy = data1_lt.copy()
data1_lt_copy =data1_lt_copy[data1_lt_copy.三级分类.isin(['台式机', '笔记本', '一体机', '平板电脑', '打印机', '多功能一体机', '复合机', '投影机', '扫描设备', '路由器', '交换机', '空调', '平板电视', '洗衣机', '冰箱'])]
```

```python
data1_lt_copy['date'] = pd.to_datetime(data1_lt_copy['订单时间'],infer_datetime_format=True)  
data1_lt_pivot = pd.pivot_table(data1_lt_copy, index=['date'], columns=['三级分类'], values=['金额'], aggfunc = np.sum)
data1_lt_pivot = data1_lt_pivot.reset_index()
data1_lt_pivot = data1_lt_pivot.fillna(0)
data1_lt_pivot_copy = data1_lt_pivot.copy()
data1_lt_pivot_copy['date'] = [datetime.strftime(x, '%Y/%m/%d') for x in data1_lt_pivot_copy['date']]
data1_lt_pivot_copy_2 =data1_lt_pivot_copy.loc[:,['金额', 'date']]
data1_lt_pivot_copy_2 = data1_lt_pivot_copy_2.set_index('date')
data1_lt_pivot_copy_2 = data1_lt_pivot_copy_2.sort_index()
data1_lt_pivot_copy_2 = data1_lt_pivot_copy_2.reset_index()
```

```python
all_in_one = []
interchanger = []
freezer = []
desktop = []
compound = []
MFP = []
tablet_pc = []
tablet_tv = []
printer = []
scanning = []
projector = []
washing = []
air_condition = []
notebook = []
router =[]
mon = []
year = []

for item in data1_lt_pivot.index:
    if data1_lt_pivot['date'][item].year == 2017:
        for month in range(1,13):
            if data1_lt_pivot['date'][item].month == month:
                mon.append(month)
                all_in_one.append(data1_lt_pivot['一体机'][item])
                interchanger.append(data1_lt_pivot[ '交换机'][item])
                freezer.append(data1_lt_pivot[ '冰箱'][item])
                desktop.append(data1_lt_pivot['台式机'][item])
                compound.append(data1_lt_pivot['复合机'][item])
                MFP.append(data1_lt_pivot['多功能一体机'][item])
                tablet_pc.append(data1_lt_pivot[ '平板电脑'][item])
                tablet_tv.append(data1_lt_pivot['平板电视'][item])
                printer.append(data1_lt_pivot[ '打印机'][item])
                scanning.append(data1_lt_pivot['扫描设备'][item])
                projector.append(data1_lt_pivot[ '投影机'][item])
                washing.append(data1_lt_pivot['洗衣机'][item])
                air_condition.append(data1_lt_pivot[ '空调'][item])
                notebook.append(data1_lt_pivot['笔记本'][item])
                router.append(data1_lt_pivot['路由器'][item])
                year.append(2017)
                
    if data1_lt_pivot['date'][item].year == 2018:
        for month in range(1,13):
            if data1_lt_pivot['date'][item].month == month:
                mon.append(month)
                all_in_one.append(data1_lt_pivot['一体机'][item])
                interchanger.append(data1_lt_pivot[ '交换机'][item])
                freezer.append(data1_lt_pivot[ '冰箱'][item])
                desktop.append(data1_lt_pivot['台式机'][item])
                compound.append(data1_lt_pivot['复合机'][item])
                MFP.append(data1_lt_pivot['多功能一体机'][item])
                tablet_pc.append(data1_lt_pivot[ '平板电脑'][item])
                tablet_tv.append(data1_lt_pivot['平板电视'][item])
                printer.append(data1_lt_pivot[ '打印机'][item])
                scanning.append(data1_lt_pivot['扫描设备'][item])
                projector.append(data1_lt_pivot[ '投影机'][item])
                washing.append(data1_lt_pivot['洗衣机'][item])
                air_condition.append(data1_lt_pivot[ '空调'][item])
                notebook.append(data1_lt_pivot['笔记本'][item])
                router.append(data1_lt_pivot['路由器'][item])
                year.append(2018)
                
data_pivot_day = pd.DataFrame(list(zip(mon, all_in_one, interchanger, freezer, desktop, compound, MFP, tablet_pc, tablet_tv, printer, scanning, projector, washing, air_condition, notebook, router, year)))
data_pivot_day.columns = ['月份', '一体机', '交换机', '冰箱', '台式机', '复合机', '多功能一体机', '平板电脑', '平板电视', '打印机', '扫描设备', '投影机', '洗衣机', '空调', '笔记本', '路由器', '年份']
```

```python
#生成按天计算的透视表的日期列
data_pivot_day['date'] =1
#将年月合并到一列
for i in range(0,len(data_pivot_day)):
    data_pivot_day['date'][i] = '%s'%data_pivot_day['年份'][i] + '/' + '%s'%data_pivot_day['月份'][i]
    
data_pivot_month = pd.pivot_table(data_pivot_day, index=['date'] ,values=['一体机', '交换机', '冰箱', '台式机', '复合机', '多功能一体机', '平板电脑', '平板电视',
       '打印机', '扫描设备', '投影机', '洗衣机', '空调', '笔记本', '路由器'], aggfunc = np.sum)
data_pivot_month = data_pivot_month.reset_index()
data_pivot_month['date'] = [datetime.strptime(x, '%Y/%m') for x in data_pivot_month['date']]
data_pivot_month.sort_values("date")
data_pivot_month_copy = data_pivot_month.copy()
data_pivot_month_copy  = data_pivot_month_copy.set_index('date')
data_pivot_month_copy  = data_pivot_month_copy.sort_index()
```
